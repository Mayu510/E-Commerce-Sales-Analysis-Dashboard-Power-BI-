```python
import pandas as pd

df = pd.read_csv (r"C:\Users\mahes\Desktop\customer_shopping_behavior.csv")
```

```python
df.isnull().sum()
```

```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median())) #to perform median category wise and fill nullvalue
```

```python
df.isnull().sum() # checking is any null values 
```

```python
df.columns = df.columns.str.lower()   # using snake casing (removes blank spaces and making words into lower cases)
df.columns = df.columns.str.replace(" " , "_")
df.columns = df.columns.str.replace("purchase_amount_(usd)", "purchase_amount")
```

```python
df.isnull().sum()
```

```python
# Creating a column
labels = ["Young Adult" , "Adult" , "Middle Age", "Senior"]
df["age_group"] = pd.qcut(df["age"] , q = 4 , labels = labels)
```

```python
df[["age" , "age_group"]].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
'Fortnightly': 14,
'Weekly': 7,
'Monthly': 30,
'Quarterly': 90,
'Bi-Weekly': 14,
'Annually': 365,
'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied' , 'promo_code_used']].head(10)
```

```python
(df["discount_applied"] == df["promo_code_used"]).all() # means both has same values
```

```python
df = df.drop("promo_code_used", axis = 1) #axis 1 means remove the column
```
